```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('shopping.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
#taking two rows
data=data.iloc[:,3:5]
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title("Customer Dendogram")
dend=shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
cluster.fit_predict(data)
```

```python
cluster.labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
data=pd.read_csv("Wholesale customers data.csv")
```

```python
data.head()
```

```python
from sklearn.preprocessing import normalize
data_scale=normalize(data)
```

```python
data_scale
```

```python
data=pd.DataFrame(data_scale,columns=data.columns)
```

```python
data.head()
```

```python
plt.figure(figsize=(10,7))
plt.title("Customer Dendogram")
dend=shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
cluster.fit_predict(data)
```

```python
cluster.labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:,3],data.iloc[:,4],c=cluster.labels_,cmap='rainbow')
```
