## 1) Build a bar plot for contract column

```python
data=customer_data['Contract'].value_counts()
print(data)
```

```python
x=data.keys().tolist()
print(x)
```

```python
y=data.values.tolist()
print(y)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fig =plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.bar(x,y,color="orange")
ax.set_xlabel("Contract type of the Customer")
ax.set_ylabel("Count")
ax.set_title("Distribution of Contract")
```

## 2) histogram for monthly chargees column

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
bins=np.arange(0,120,10)
ax.hist(customer_data['MonthlyCharges'],bins,color="forestgreen")
ax.set_xlabel("Monthly charges incurred")
ax.set_ylabel("Count")
ax.set_title("Distribution of Monthly ranges")
ax.set_xticks(bins)
ax.set_yticks(np.arange(0,1601,100))
ax.grid(True)
ax.legend(labels=['1'],loc=2)
```

# 3) Scatter plot between total charges and tenure

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
ax=fig.add_axes([0,0,2,2])
ax.scatter(customer_data['tenure'],customer_data['TotalCharges'],color='indigo')
ax.set_xlabel("Tenure")
ax.set_ylabel("Total Charges")
ax.set_title("TotalCharges vs Tenure")
ax.set_xticks(np.arange(0,90,10))
```
