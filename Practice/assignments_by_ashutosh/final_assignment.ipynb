
1.create histogram for time spend on day calls by customer and find what is the day minutes range where maximum customers are using it

```python
#)sol- (150 - 200)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
churn=pd.read_csv(r"C:\Users\Lenovo\Downloads\churn.csv")
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
binx=np.arange(0,351,50)
ax.hist(churn['total day minutes'],binx)
ax.set_title('Number of customers spending in the range total day minutes')
ax.set_xlabel('total day minutes')
ax.set_ylabel('Customers')
ax.set_xticks(binx)
```

2.categorize the churner and non-churner for the time spent on day calls

```python
#)sol
import matplotlib.pyplot as plt
import numpy as np
churn=pd.read_csv(r"C:\Users\Lenovo\Downloads\churn.csv")
churn1=churn[churn["churn"]]
churn2=churn[(churn["churn"]==False)]
sum1=churn1["total day calls"].sum()
sum2=churn2["total day calls"].sum()
a=np.array([sum1,sum2])
Category=["Churner","Non-Churner"]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(Category,a,width=0.5)
ax.set_title("Total day calls by Churner and Non Churner")
ax.set_xlabel("Category")
ax.set_ylabel("Total Day Calls Spent")
```
