# Two types of plotting

## 1) using figure and adding axis and then plotting


import matplotlib.pyplot as plt

fig=plt.figure()


```python
import matplotlib.pyplot as plt
import numpy as np
import math
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
ax.plot(x,y)
ax.set_title('Figure')
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```

## 2) Directly plotting from mathplotlib.pyplot

```python
# this will have less features compared to figure
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.title('Figure')
plt.xlabel('angle')
plt.ylabel('sine')
plt.show()
```

# BAR CHART

```python
div=["A","B","C","D","E"]
av_marks=[40,50,70,60,30]
plt.bar(div,av_marks,color='lightgreen')
plt.xlabel('Divisions')
plt.ylabel('Marks')
plt.title('Barchart')
```

```python
import matplotlib.pyplot as plt
import numpy as np
div=np.arange(2014,2019)
av_marks=[40,50,70,60,30]

fig=plt.figure()
ax=fig.add_axes([0,0,3,1])
ax.bar(div,av_marks,color="orange",width=0.25)
ax.set_title('Category Wise Marks')
ax.set_xlabel('Category')
ax.set_ylabel('Marks')

av_marks1=[30,40,60,20,50]
ax.bar(div+0.25,av_marks1,color="green",width=0.25)
ax.legend(['cs','ece'],loc=2)
```

# PIE CHART

```python
div=["A","B","C","D","E"]
shares=[20,25,15,10,30]
eplode=[0,0.1,0,0.5,0]
plt.pie(shares,labels=div,explode=eplode,shadow=True)
#plt.gca().legend(div,loc="lower right")
plt.legend(loc="lower right",title="DIVS")
```

# HISTOGRAM

```python
x=[5,5,10,10,21,21,21,21,21,50,100,51,87,84,89,90,58]
binx=np.arange(0,101,25)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.hist(x,binx,facecolor='darkgreen',alpha=0.9)
ax.set_xticks(np.arange(0,101,25))
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins = [0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
```
